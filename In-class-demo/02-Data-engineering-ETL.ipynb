```python
import pandas as pd
```

```python
data = pd.read_excel("datasets/01-attendantsheet-sample.xlsx")
```

```python
data.head(30)
```

```python
start_row = 0
end_row = 0
end_column = 0
```

```python
total_rows, total_columns = data.shape
total_rows, total_columns
```

```python
for row in range(total_rows):
    sequence_num = data.iloc[row, 0]
    print(sequence_num, type(sequence_num))
```

```python
for row in range(total_rows):
    sequence_num = data.iloc[row, 0]
    if "Date".lower() in str(sequence_num).lower():
        print(f"Found Start Row: {row}")
        start_row = row
    if "legend".lower() in str(sequence_num).lower():
        print(f"Found End Row: {row}")
        end_row = row
```

```python
filtered_row_data = data.iloc[start_row:end_row]
filtered_row_data.head(5)
```

```python
filtered_row_data.head(2).T
```

```python
total_columns
```

```python
for column in range(total_columns):
    column_header = filtered_row_data.iloc[0, column]
    if str(column_header).lower() == "total".lower():
        print(f"Found End Column: {column}")
        end_column = column
```

```python
filtered_row_data
```

```python
attendant_datasets = filtered_row_data.iloc[:, :end_column]
attendant_datasets = attendant_datasets[attendant_datasets.iloc[:, 0].notna()]
attendant_datasets.head()
```
