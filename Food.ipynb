<b>Exercise 5. Display the <i>"food"</i> which has the highest number of orders?

```python
food_counts = order_list['food'].value_counts()
most_ordered_food = food_counts.idxmax()
print(f"The food with the highest number of orders is: {most_ordered_food}")
print(f"Number of orders: {food_counts[most_ordered_food]}")
```

<b>Exercise 5. How many items were ordered for the most-ordered item?

```python
most_ordered_food_count = food_counts[most_ordered_food]
print(f"Number of items ordered for the most-ordered food: {most_ordered_food_count}")
```

<b>Exercise 6. Display the <i>"description"</i> which has the highest number of orders?</b>

```python
description_counts = order_list['description'].value_counts()
most_ordered_description = description_counts.idxmax()
print(f"The description with the highest number of orders is: {most_ordered_description}")
print(f"Number of orders: {description_counts[most_ordered_description]}")
```

<b>Exercise 7. How many foods ordered in total?</b>

```python
total_foods = order_list['food'].value_counts().sum()
print(f"Total number of foods ordered: {total_foods}")
```

<b>Exercise 8. Calculate the total income for the period in the dataset?</b>

```python
order_list['unit_price'] = order_list['unit_price'].str.replace('$', '', regex=False).astype(float)
order_list['line_total'] = order_list['unit_price'] * order_list['quantity']
total_income = order_list['line_total'].sum()
total_orders = order_list['id'].nunique()
average_order_value = total_income / total_orders
print(f"Total Income: ${total_income:,.2f}")
print(f"Total Unique Orders: {total_orders}")
print(f"Average Spend per Order: ${average_order_value:,.2f}")
```

<b>Exercise 9. Find the total number of orders.</b>

```python
total_orders = order_list['id'].nunique()
print(total_orders)
```

<b>Exercise 10. Find the average income for each order?</b>

```python
order_list['line_total'] = order_list['quantity'] * order_list['unit_price']
order_totals = order_list.groupby('id')['line_total'].sum()
average_order_value = order_totals.mean()
print(f"The average income for each order is: ${average_order_value:,.2f}")
```

<b>Exercise 11. How many different foods are ordered?</b>

```python
unique_foods = order_list['food'].nunique()
print(f"Number of unique foods ordered: {unique_foods}")
```
