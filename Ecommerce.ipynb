
5. How many rows and columns are there in our Dataset?

```python
print(df.shape)
rows, columns = df.shape
print(f"Number of rows: {rows}")
print(f"Number of columns: {columns}")
```


6. Highest and Lowest Purchase Prices.

```python
highest_price = df['Purchase Price'].max()
lowest_price = df['Purchase Price'].min()
print(f"Highest Purchase Price: {highest_price}")
print(f"Lowest Purchase Price: {lowest_price}")
```


7. Average Purchase Price

```python
avg_price = df['Purchase Price'].mean()
print(f"Average Purchase Price: {avg_price}")
```


8. How many people have French 'fr' as their Language?

```python
row, _= df[df["Language"] == 'fr'].shape
row
```

9. How many languages people have in total

```python
total_languages = df['Language'].nunique()
print(f"Total number of languages: {total_languages}")
```


9. Job Title Contains Engineer or engineer

```python
engineer_count = df[df['Job'].str.contains('engineer', case=False)].shape[0]
print(f"Number of people with 'engineer' in their job title: {engineer_count}")
```

How many number of people have job as manager?

```python
manager_count = df[df['Job'].str.contains('manager', case=False)].shape[0]
print(f"Number of people with 'manager' in their job title: {manager_count}")
```


10. Find The Email of the person with the following IP Address: 132.207.160.22

```python
target_ip = '132.207.160.22'
target_email = df[df['IP Address'] == target_ip]['Email'].values[0]
print(f"Email of the person with IP address {target_ip}: {target_email}")
```


11. How many People have Mastercard as their Credit Card Provider and made a purchase above 50?

```python
mastercard_count = df[(df['CC Provider'] == 'Mastercard') & (df['Purchase Price'] > 50)].shape[0]
print(f"Number of people with Mastercard as their credit card provider and purchase price above 50: {mastercard_count}")
```


12. Find the email of the person with the following Credit Card Number: 4664825258997302

```python
email_cc_number = df[df['Credit Card'] == 4664825258997302]['Email'].values[0]
print(f"Email of the person with Credit Card Number 4664825258997302: {email_cc_number}")
```


13. How many people purchase during the AM and how many people purchase during PM?

```python
df['AM'] = df['AM or PM'].str.contains('AM')
df['PM'] = df['AM or PM'].str.contains('PM')
am_count = df['AM'].sum()
pm_count = df['PM'].sum()
print(f"Number of people who made purchases during the AM: {am_count}")
print(f"Number of people who made purchases during the PM: {pm_count}")
```


14. How many people have a credit card that expires in 2020?

```python
expired_2020 = df[df['CC Exp Date'].str.endswith('20')]
print(f"Number of people with credit cards that expire in 2020: {len(expired_2020)}")
```


15. What are the top 5 most popular email providers (e.g. [gmail.com](http://gmail.com/), [yahoo.com](http://yahoo.com/), etc...)

```python
domains = df['Email'].str.split('@').str[1]
top_5_providers = domains.value_counts().head(5)
print(top_5_providers)
```
